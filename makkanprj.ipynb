```python

import time
import requests
from IPython.display import clear_output
from bs4 import BeautifulSoup
import pandas as pd

base_url = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page='

own = []   
bhk = []
prz = []
sqt = []
loc = []
stat = []
pps = []

for i in range(1,11):
  time.sleep(3)

  print(f'Going to scrape data from {i} page')
  clear_output(wait=True)
  url = base_url+str(i)
  req = requests.get(url)
  soup = BeautifulSoup(req.text,'html')

  #owner info
  ow_info = soup.findAll('span',attrs={'class':'seller-type'})
  for j in ow_info:
    own.append(j.text)
  
  #bhk info
  bhk_info = soup.findAll('div',attrs={'class':'title-line'})
  for j in bhk_info:
    bhk.append(j.strong.span.text)

  #price info
  p_info = soup.findAll('div',attrs={'data-type':'price-link'})
  for  j in p_info:
    prz.append(j.text)
  
  #size info
  s_info = soup.findAll("td",attrs={'class':'size'})
  for j in s_info:
    sqt.append(j.text)
  
  #location info
  loc_info = soup.findAll('span',attrs={'itemprop':'addressLocality'})
  for j in loc_info:
    loc.append(j.text)
  
  #status info
  st_info = soup.findAll('tr',attrs={'class':'hcol w44'})
  for j in st_info:
    stat.append(j.text)

  #price per sqft info
  pps_info = soup.findAll('td',attrs={'class':'lbl rate'})
  for j in pps_info:
    pps.append(j.text)
```

```python
data = pd.DataFrame({'Owner_inf' :own ,
              'Bhk_info' : bhk,
              'Price' : prz,
              'Locality' : loc,
              'Area_sqft' : sqt,
              'Status' :stat ,
              'Price_Per_sqft' : pps})
```

```python
data.to_csv('apartmentdata.csv',index=False)
data
```
